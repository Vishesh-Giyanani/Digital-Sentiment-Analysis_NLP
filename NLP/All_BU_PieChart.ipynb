```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#reading the input file
df=pd.read_csv("Iteration_2.csv")
```

```python
# Calculate the counts for each category
counts = df['Appliances_label'].value_counts()

# Calculate the total number of responses
total = counts.sum()

# Create a pie chart
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=["lightgreen","gold","salmon"])

# Add the total as a text label in the bottom right corner
plt.text(1, -1, f'Total: {total}', ha='right', va='bottom', fontsize=12)

# Set the title and labels
plt.title('Appliances')

# Show the chart
plt.show()
```
