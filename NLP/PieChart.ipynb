```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#reading the input file
df=pd.read_csv("Final2.csv")
```

```python
# Calculate the counts for each category

counts = df['Appliances_label'].value_counts()

# Create a pie chart
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=["darkblue","orange","grey","maroon"])

# Set the title and labels
plt.title('Appliances')

# Show the chart
plt.show()
```
