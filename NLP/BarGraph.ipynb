```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("Final2.csv")
```

```python
# plot the bar chart


sns.barplot(x='Appliances_label', y = df['Unnamed: 0'], palette = ["darkblue","orange","grey","maroon"], ci = None, data=df)

plt.xticks(rotation=45)
plt.xlabel('Labelling of customers', fontsize=14)
plt.ylabel('No. of responses', fontsize=14)
plt.title("Appliances")
plt.show()
```
